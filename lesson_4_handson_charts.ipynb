#### Part 2: Create a Bar Graph 
Import 'cars' data set and view

```python
cars = pd.read_csv("C:/Users/desja/python_course/datasets/cars.csv")
cars
```

#### Part 3 Create a Stacked Bar Chart

```python
car_location = pd.read_csv("C:/Users/desja/python_course/datasets/car_location.csv")
car_location
```

#### Part 4: Create a Scatter Plot
Import 'crocodiles' data set and view

```python
crocodiles = pd.read_csv("C:/Users/desja/python_course/datasets/crocodiles.csv")
crocodiles
```

```python
Plot = crocodiles.plot.scatter(x='HeadLength', y='BodyLength')
Plot.set_title("Head Length vs Body Length of Estuarine Crocodiles")
Plot.set_xlabel("Head Length")
Plot.set_ylabel("Body Length")
```

#### Scatter Plot shows a right upward trend: 
meaning that as the head length of the crocodile increases, the body length also tends to increase. 
This indicates a positive correlation between the two variables.
